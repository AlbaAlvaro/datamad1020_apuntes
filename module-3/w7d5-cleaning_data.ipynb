```python
df["PrecipitationSumInches"] = df["PrecipitationSumInches"].str.replace("T","-")
columns = ['DewPointAvgF', 'DewPointLowF', 'HumidityHighPercent',
       'HumidityAvgPercent', 'HumidityLowPercent',
       'SeaLevelPressureHighInches', 'SeaLevelPressureAvgInches',
       'SeaLevelPressureLowInches', 'VisibilityHighMiles',
       'VisibilityAvgMiles', 'VisibilityLowMiles', 'WindHighMPH', 'WindAvgMPH',
       'WindGustMPH',"PrecipitationSumInches"]
```

```python
missing_ind = {}
for col in columns:
    missing_ind[col] = df[df[col] == "-"].index
```

```python
lst = list(missing_ind['DewPointAvgF'])
```

```python
[[174, 175, 176, 177],[596, 597, 598]]
```

```python
def split_sequences(lst):
    res = []
    seq = []
    for prev,post in zip(lst,lst[1:]):
        seq.append(prev)
        if post != prev + 1:
            res.append(seq)
            seq=[]
    seq.append(post)
    res.append(seq)
    return res
```

```python
list(zip(lst,lst[1:]))
```

```python
for col, values in missing_ind.items():
    missing = split_sequences(values)
    for seq in missing:
        mini,maxi = min(seq),max(seq)
        mini -= 1
        maxi += 1
        df[col].iloc[seq] = (float(df[col].iloc[mini])+float(df[col].iloc[maxi]))/2
    df[col] = df[col].astype(float)
```
